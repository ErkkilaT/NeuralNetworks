### Introduction
The aim of this project was to sort pictures of hands holding up 2 or 3 fingers and try to find a model that can separate them effectively. 
### The Dataset
The dataset consist of 128 (64 of each) pictures of hands holding up 2 or 3 fingers. They were self taken on a mobile phone and then resized to 180x180 pixels. The pictures include the hands of 2 different people so it is good to recognize that this will set inherent limitations on any model built. To compensate somewhat, different ways of holding up fingers(holding different sets of fingers) were used and some variations(big rings on fingers etc.) were added.
# The models
Three separate models were built. The code, some training data and visualizations for each model can be found in their respective documentations. Below will be general results/analysis/comparisons of the models.
 - One from scratch with a self-trained convolutional network.
 - Second using VGG16 for feature extraction just adding some fully connected layers to the end.
 - Third using VGG16 but finetuning the last few layers.  
These models will be referred to as **CNN_model**, **feature_extraction_model** and **fine_tuning_model** respectively.

### Model accuracies
| Model | Accuracy |
|-----------------:|:---------------|
|                CNN_model|                0.7500|  
| feature_extraction_model|                0.8500|   
|        fine_tuning_model|                0.8500|    

 

The performance of all three models is good against a dummy-model(50%) with the two models using VGG16 being better than the one made from scratch (85% vs 75% accuracy). The dataset being so small in addition to making training harder also made testing the model more difficult. During my testing the feature extraction version seemed to generally perform best, but since there are only 20 pictures in the test set it became hard to differentiate between it and the fine tuning -model since both got generally 85% or 90% with different setups/runs. The fine tuned one gave similar accuracies with most setups but was more erratic than the feature extraction -model.  
### Model tuning challenges
For the from scratch -model I had problems where the model easily just reverted to ~50% accuracy with even quite slight changes. Ended up with a very small model because adding more/bigger convolutional layers did not improve the model. Also adding a dense layer at the end (which i've found worked well in the cifar-10 project) basicly broke the model making it just guess. Most of these problems seem to stem from the dataset being small. I would also guess that the fact that the picture contain a lot of similarities(background and the palm/arm part of the hand are very similar in all pictures) is the reason for the extra dense layer being so bad for the model.  

The feature extraction version was quite easy to tune since we're only adjusting the end of the model. Tried adding more layers/cells but that didn't seem to improve the model at all so ended up with a single 128 cell FCN-layer.  

Initially the fine tuning version was very difficult to get to do anything but ~50% accuracy(dummy) because i started with the last 4 layers of VGG16 open, but just leaving the last 2 open made the model work well. After the VGG16 part, the rest is basicly the same as the feature extraction version.
### Other challenges
I initially planned to do this project with pictures of knives/forks/spoons but realized that I took the pictures on way too different backgrounds for any of the versions to get over 33% (which is no better than guessing). Then pivoted to the pictures of fingers with the limitation of doing it last minute so only being able to use a few persons hands.  

Also struggled with a weird interaction importing the pictures from phone->windows->Linux where the orientation of the photos got mangled. Spent a lot of time troubleshooting my data_augmentation function before realizing what the actual problem was. Luckily there was a ready made tool in Python Imaging Library to apply the correct orientation from the Exif-data to the actual picture(in CNN_project_2 -file).
### Model application
Since the dataset used was so limited in just raw amount of data and the similarity of the pictures, any of the models would likely yield bad results if not used strictly for similar pictures(same zoom/background/etc). If you had to pick from the models made the two latter ones would probably be best for recognizing numbers fingers in pictures. If you kept developing the model the results of the feature extraction -model seemed more easily improvable.
